## 3. Final Correct Code

```python
class Solution(object):
  def isValid(self, s: str) -> bool:

    # dictionary to show which open matches with closed
    pairs = {
        '(': ')',
        '{': '}',
        '[': ']'
    }

    # stack to hold open brackets
    stack = []

    # traversal through string, its characters
    for char in s:

      if char in "([{": # if char is open bracket push to stack
        stack.append(char)

      else:

        if len(stack) == 0: # invalid if closing bracket and stack empty
          return False

        # pop most recent open bracket
        last = stack.pop()

        # a check to see if brackets match
        if char != pairs[last]: # Corrected line: compare char with the expected closing bracket for 'last'
          return False

    return len(stack) == 0 # return true if stack empty

# user inputs
userStr = (input("Enter string of brackets: "))

# instance of solution
sol = Solution()

# call method
result = sol.isValid(userStr)
print("The string you have entered is:", result)

# Used Gemini with prompt ‚ÄúHow to write integrate dictionary properly‚Äù
```

```python
# Final working solution

class Solution(object):
  def countStudents(self, students, sandwiches):

    # count students that want 0 or 1
    count_zero = 0
    count_one = 0

    for s in students:
      if s == 0:
          count_zero += 1
      else:
          count_one += 1

    # going through each sandwich
    for sandwich in sandwiches:
      if sandwich == 0:

        # stop if no one wants
        if count_zero == 0:
          break
        count_zero -= 1

      else: # sandwich == 1

        if count_one == 0:
          break
        count_one -= 1

    return count_zero + count_one # Added return statement

# user inputs
userStu_str = input("Enter student preferences (e.g., 0,1,0): ")
students_list = [int(x) for x in userStu_str.split(',')]

userSand_str = input("Enter sandwich types (e.g., 0,1,1): ")
sandwiches_list = [int(x) for x in userSand_str.split(',')]

# instance of solution
sol = Solution()

# call method
result = sol.countStudents(students_list, sandwiches_list)
print("Number of students who did not eat:", result)

# Used Gemini with prompt ‚ÄúHow to integrate the user input for their int letting it iterate properly‚Äù
```

```python
# Final working solution

class Solution(object):
  def isAnagram(self, s: str, t: str) -> bool:

    # lengths differ false
    if len(s) != len(t):
      return False

    # converting strings to lists
    s_list = list(s)
    t_list = list(t)

    # selection sort (used Colab autofill)
    def selection_sort(arr):
      n = len(arr)
      for i in range(n):
        min_index = i
        for j in range(i+1, n):
          if arr[j] < arr[min_index]:
            min_index = j

        # swapping into correct position
        arr[i], arr[min_index] = arr[min_index], arr[i]

    # sorting both lists
    selection_sort(s_list)
    selection_sort(t_list)

    # comparing and returning
    return s_list == t_list

# user inputs
userStr1 = (input("Enter your first word: "))
userStr2 = (input("Enter your second word: "))

# instance of solution
sol = Solution()

# call method
result = sol.isAnagram(userStr1, userStr2)
print("Are your words anagram of each other?", result)

# Used Gemini with prompt "How to know when to swap elements into correct position‚Äù
```
