```python
import gradio as gr

def binary_search_visual(arr_string, target_string):

  # validation
  # split into spaces, convert string to int with map, then make list
  try:
    arr = list(map(int, arr_string.split()))
    target = int(target_string)

  except:
    return "Please enter valid integers only."





  # ensuring the list is sorted since binary requires sorted
  arr_sorted = sorted(arr)

  steps = [] # empty list stores the steps of 2 pointer search
  left, right = 0, len(arr_sorted) - 1 # left and right pointers
  step_num = 1 # keeps track of which step user on





  # starting the actual searching process
  steps.append(f"Starting Binary Search on sorted list: {arr_sorted}\n")

  while left <= right:
    mid = (left + right) // 2 # middle index
    mid_value = arr_sorted[mid] # take value stored at middle index

    steps.append(f"Step {step_num}:")
    steps.append(f"Left Pointer: {left}")
    steps.append(f"Right Pointer: {right}")
    steps.append(f"Middle Index: {mid}")
    steps.append(f"Middle Value: {mid_value}")

    # checking if value matches
    if mid_value == target:
      steps.append(f"Target {target} found at index {mid}!")
      return "\n".join(steps) # all strings into one big string with seperation

    # if target smaller search left
    elif target < mid_value:
      steps.append(f"Target {target} < {mid_value}, searching the left half.\n")
      right = mid - 1

    # if target larger search right
    else:
      steps.append(f"Target {target} > {mid_value}, searching the right half.\n")
      left = mid + 1

    step_num += 1 # counter

  # if loop ends with no result then
  steps.append(f"Target {target} not found in the list.")
  return "\n".join(steps) # all strings into one big string with seperation





# Gradtio implementations now
# making UI for binary search function
interface = gr.Interface(
    fn=binary_search_visual, # function that will be called when user hits submit

    # the inputs, 1st one is textbox for list of numbers, 2nd is the target number
    inputs=[
        gr.Textbox(
            label = "Enter a list of numbers separated by spaces",
            placeholder = "E.g. 1 2 3 4 5 6 7"
        ),
        gr.Textbox(
            label = "Enter the target number",
            placeholder = "E.g. 7"
        )
    ],

    # output for user to see
    outputs=gr.Textbox(
        label="Binary Search Steps",
        lines=15,
        placeholder="Output will show here."
    ),

    # title
    title="A Visual Walkthrough of Binary Search",

    # short desc. of how app works
    description="Enter list of numbers and your desired target. The app will now show each binary step visually.",

    # changing the theme color
    theme=gr.themes.Ocean()
)

# launching the app
interface.launch()
```
